```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
print(py_st_age)
```

```python
plt.hist(m1_st_age)
plt.title('ML student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show()
```

```python
# here the bars are not showing properly. lets give another parameter
```

```python
bins =[15,20,25,30,35,40,45]
plt.hist(m1_st_age,bins, rwidth = 0.8,histtype = "step")
plt.title('ML student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show() 
# histogram in the form of step
```

```python
bins =[15,20,25,30,35,40,45]
plt.hist(m1_st_age,bins, rwidth = 0.6,histtype = "bar") # in the formof bar
plt.title('ML student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show() 
```
