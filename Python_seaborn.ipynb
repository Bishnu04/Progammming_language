# how to draw histogram by using distplot function

```python
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset("tips")
tips_df # loaded data from github
```

```python
# only few are numeric
```

```python
sns.distplot(tips_df["size"])
```

```python
# we have histogram with a line plot
# kde kernal density estimtor 
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["total_bill"])
```

```python
# use the bins parameter.
# suppose i want to print the data in the range of 5 or 10
```

```python
sns.distplot(tips_df["total_bill"],bins=10)
```

```python
# lets provide the bin number 100
```

```python
sns.distplot(tips_df["total_bill"],bins=100)
```

```python
# next parameter is hist default is true
```

```python
sns.distplot(tips_df["total_bill"],bins=100,hist=False)
```

```python
# just the KDE kernal density estimator or line plot
# the histogram can be hide this way
```

```python
# next parameter is 
```

```python
sns.distplot(tips_df["total_bill"],bins=100,kde=False)
```

```python
# This way got the hsitogram only not the kde
```

```python
# next parameter is the rug plot, default value is false
```

```python
sns.distplot(tips_df["total_bill"],rug=False)
```

```python
sns.distplot(tips_df["total_bill"],rug=True)
```

```python
# where  the data is present, there are some bars on the x axis
```

```python
sns.distplot(tips_df["total_bill"],fit=norm)
```

```python
# there is another line plot.
# if you need only one do kd = false
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,kde=False)
```

```python
# color parameter
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,kde=False,color ='g')
```

```python
# next parameter is vertical and default value is false
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,kde=False,color ='g', vertical = True)
```

```python
# next parameter is norm hist
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,kde=False,color ='g',norm_hist=True)
```
