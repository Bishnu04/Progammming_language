# difference between view and copy

```python
import numpy as np
arr = np.array([5,10,15,20,25,30]) # i will make a slice of this  1d array by making a new variable
print(arr)
slice1=arr[2:5]
print(slice1) # now I will make the element of the slice1 0 by the following way
slice1[:]=0
print(slice1) # now lets check the array itself
print(arr) # the result  is three middle elements are zero
# the reason is the slice1 is the view of the original that  is they have the same memory location
```

```python
#in numpy you can do the broadcasting
```

```python
import numpy as np
arr = np.array([5,10,15,20,25,30]) # i will make a slice of this  1d array by making a new variable
print(arr)
slice1=arr[2:5].copy()
print(slice1) # now I will make the element of the slice1 0 by the following way
slice1[:]=0
print(slice1) # now lets check the array itself
print(arr)
# the reason is the slice1 is the copy of the original that means they have the different memory location
```

```python
#lets understand the parameters of the dropna() method
# put curser inside the parenthesis and then press shift and tab
# the first parameter is axis which removes the rows or columns with missing values
```

```python
# for axis parameter we provide 0 or 1
# 0 means row  or index and 1 means column. It has a default value is 0.
# letsput axis = 1 and how it looks like
```
